```python
with open('stopwords.txt') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n', '') for line in stopwords]
stopwords
```

```python
import nltk
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


nltk.download('punkt_tab')     
nltk.download('stopwords')

stopwords = set(stopwords.words('english'))

# ' Question'
dataset = pd.DataFrame(columns=['title_body'])

for index, row in X.iterrows():
    title_body_tokenized = word_tokenize(row[' Question'])
    title_body_tokenized_filtered = [w.lower() for w in title_body_tokenized if w.lower() not in stopwords]
    
    # ( raw string)
    s = re.sub(r'[^\w\s]', '', ' '.join(title_body_tokenized_filtered))
    s = re.sub(r"\d+", "", s)
    
    dataset.loc[index] = {'title_body': s}
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(dataset['title_body'])
```

```python
X = vectorizer.transform(dataset['title_body'])
```
